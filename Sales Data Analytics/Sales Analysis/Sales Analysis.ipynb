```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None



# Read in the data
df = pd.read_csv(r'/Users/nani/Documents/DA PORTFOLIO/AtliQ Sales data/transactions.csv')
```

```python
# Head of the data
df.head()
print(df.dtypes)
```

```python
# Correlation plot between the cost_price and the sales_amount
sns.regplot(x=df['cost_price'],y=df['sales_amount'])
```

```python
# From the plot there is a high Correlation between the Sales_amount and the Cost_Price 
 
df.corr(method='pearson')
```

```python
# Correlation value is very high (0.941800)
# Concluded that most of the products are sold for a profit

correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot = True)
plt.show()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
```

```python
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot = True)
plt.show()
```
